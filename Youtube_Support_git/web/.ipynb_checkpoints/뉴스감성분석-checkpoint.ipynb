```python
from konlpy.tag import Kkma,Komoran,Hannanum,Okt,Mecab
import pandas as pd
import re

#mecab = Mecab()
kkma = Kkma()
okt = Okt()
```

```python
train_df = pd.read_csv('./nsmc/ratings_train.txt',sep='\t',encoding='utf8')
train_df
```

```python
train_df = train_df.fillna(' ')
train_df.info()
```

```python
train_df['document'] = train_df['document'].apply(lambda x:re.sub(r'\d+','',x))
```

```python
test_df = pd.read_csv('./nsmc/ratings_test.txt',sep='\t')
test_df = test_df.fillna(' ')
test_df['document'] = test_df['document'].apply(lambda x:re.sub(r'\d+','',x))

test_df.info()
```

```python
train_df.drop(columns=['id'],inplace=True)
test_df.drop(columns=['id'],inplace=True)
```

```python
okt = Okt()

def tokenizer(text):
    return okt.morphs(text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
tfid_vect = TfidfVectorizer(tokenizer = tokenizer,
                          ngram_range=(1,2),
                             min_df=3,
                             max_df=0.9)

tfid_matric_train = tfid_vect.fit_transform(train_df['document'])
tfid_matric_test = tfid_vect.transform(test_df['document'])
```

```python
lg_clf = LogisticRegression(random_state=0,solver='liblinear')
params={
    'C':[3.5]
}
grid_cv = GridSearchCV(lg_clf,param_grid=params,cv=3,scoring='accuracy',verbose=1)
grid_cv.fit(tfid_matric_train,train_df['label'])
grid_cv.best_score_
```

```python
pred = grid_cv.best_estimator_.predict(tfid_matric_test)
accuracy_score(test_df['label'],pred)
```

```python
#정규표현식 (...이나 이런거 사라짐)
import re

def apply_regular_expression(text):
    hangul = re.compile('[^ ㄱ-ㅣ 가-힣]')
    result = hangul.sub('', text)
    return result
```

```python
# for i in comment_df.iloc[:, 0]:
#     print(i)

pos_comment = []
neg_comment = []

for i, v in enumerate (comment_df.iloc[:, 0]):
    if comment_df.loc[i, '감상'] == '긍정':
        pos_comment.append(apply_regular_expression(v))
    elif comment_df.loc[i, '감상'] == '부정':
        neg_comment.append(apply_regular_expression(v))
        
print(len(pos_comment), len(neg_comment))
```

```python
okt_pos = []

for i in range(len(pos_comment)):
    okt_pos.append(Okt().pos(pos_comment[i], norm=True, stem=True))
okt_pos
```

```python
# 명사, 동사, 형용사, 이모티콘, 동사 접두사 사용
# noun, verb, adjective, KoreanParticle, VerbPrefix

# 한글자 형태소 제거 및 해당 형태소 단어 추출
a = []

for j in range(len(okt_pos)):
    b = []
    for i in range(len(okt_pos[j])):
        if (okt_pos[j][i][1] == 'Noun') and (len(okt_pos[j][i][0]) > 1):
            b.append(okt_pos[j][i])
        elif (okt_pos[j][i][1] == 'Verb') and (len(okt_pos[j][i][0]) > 1):
            b.append(okt_pos[j][i])
        elif (okt_pos[j][i][1] == 'Adjective') and (len(okt_pos[j][i][0]) > 1):
            b.append(okt_pos[j][i])
        elif (okt_pos[j][i][1] == 'KoreanParticle') and (len(okt_pos[j][i][0]) > 1):
            b.append(okt_pos[j][i])
        elif (okt_pos[j][i][1] == 'VerbPrefix') and (len(okt_pos[j][i][0]) > 1):
            b.append(okt_pos[j][i])
    a.append(list(b))
a
```
