```python
news['댓글 내용'] = news['댓글 내용'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")

# 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

# 형태소 분석기 OKT를 사용한 토큰화 작업 (다소 시간 소요)
okt = Okt()

news_token = []
for sentence in tqdm(news['댓글 내용']):
    tokenized_sentence = okt.morphs(sentence, stem=True) # 토큰화
    stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # 불용어 제거
    news_token.append(stopwords_removed_sentence)
news_token
```

```python
import json

# class KnuSL():

def data_list(wordname):
    with open('KnuSentiLex/data/SentiWord_info.json', encoding='utf-8-sig', mode='r') as f:
        data = json.load(f)
    result = ['None','None']
    for i in range(0, len(data)):
        if data[i]['word'] == wordname:
            result.pop()
            result.pop()
            result.append(data[i]['word_root'])
            result.append(data[i]['polarity'])

    r_word = result[0]
    s_word = result[1]
    
    if r_word is not None and s_word is not None:
        return r_word, s_word
    else:
        pass
```

```python
score_li = []
score_list = []
for i in tqdm(range(len(news_removeone))):
    #print(text_list[i])
    score_li = []
    for j in news_removeone[i]:
        #print(j)
        if data_list(j)[0] == 'None': 
            pass
        else:
            score =  (data_list(j))
            score_li.append(score)
    score_list.append(score_li)
score_list
```

```python
sco, scor = [], []
for i in range(len(score_list)):
    sco = []
    for j in range(len(score_list[i])):
        sco.append(int(score_list[i][j][1]))
    scor.append(sco)
scor
```

```python
score_sum = []
for i in range(len(scor)):
    score_sum.append(sum(scor[i]))
score_sum_df = pd.DataFrame(score_sum)
score_sum_df.columns = ['sentiment_score']
score_sum_df
```

```python
news = news.reset_index(drop = True)
```

```python
news_sentiment_score = pd.concat([news, score_sum_df], axis=1)
news_sentiment_score
```
